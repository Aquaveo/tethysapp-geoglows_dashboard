# Kenya Data Prep

```python
import pandas as pd
import geopandas as gpd
```

```python
workspace = '/home/ysun/tethys-dev/tethysapp-geoglows_dashboard/tethysapp/geoglows_dashboard/workspaces/app_workspace'
data_folder = '/home/ysun/tethys-dev/tethysapp-geoglows_dashboard/tethysapp/geoglows_dashboard/public/data/geojson'
```

## Hydro Stations

```python
df_station = pd.read_csv(f'{workspace}/kenya_stations.csv')
gdf_station = gpd.GeoDataFrame(df_station, geometry=gpd.points_from_xy(x=df_station.Longitude, y=df_station.Latitude), crs="EPSG:4326")
gdf_station = gdf_station.drop(columns=['Latitude', 'Longitude'])
gdf_station
```

```python
gdf_station.info()
```

```python
gdf_station.to_file(f'{data_folder}/kenya_stations.geojson', driver='GeoJSON')
```

## Sub-basins

```python
df_basin = pd.read_csv(f'{workspace}/kenya_streams.csv').drop(columns=['fid'])
df_basin.info()
```

```python
gdf_basin = gpd.read_file(f'{data_folder}/Kenya_sub_basins.geojson')
gdf_basin.info()
```

```python
gdf_basin_merge = gdf_basin.merge(df_basin, on='Subbasin', how='outer').drop(columns=['OBJECTID_1'])
gdf_basin_merge
```

```python
gdf_basin_merge.info()
```

```python
gdf_basin_merge.to_file(f'{data_folder}/kenya_sub_basins.geojson', driver='GeoJSON')
```
