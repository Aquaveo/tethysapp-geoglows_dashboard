```python
import pandas as pd
import geopandas as gpd
import pyogrio
```

# ID Mapping

```python
df_stream_order = pd.read_parquet("workspaces/app_workspace/v2-master-table.parquet")
df_stream_order
```

```python
df_river_id = gpd.read_file("public/data/geojson/hydrosos_streamflow_output.geojson")
df_river_id
```

```python
df_geometry = pyogrio.read_dataframe("workspaces/app_workspace/global_streams_simplified.gpkg")
df_geometry
```

```python
df = df_river_id[["rivid"]].merge(df_stream_order[["LINKNO", "strmOrder"]], how="left", left_on="rivid", right_on="LINKNO").drop("LINKNO", axis=1)
df = df.merge(df_geometry, how="left", left_on="rivid", right_on="LINKNO").drop("LINKNO", axis=1)
df
```

```python
df.info()
```

```python
df = df.dropna()
df["strmOrder"] = df["strmOrder"].astype("int")
df
```

```python
gdf = gpd.GeoDataFrame(df, geometry=df["geometry"])
gdf
```

```python
gdf.to_file("public/data/geojson/hydrosos_streamflow_geometry.geojson", driver="GeoJSON")
```
