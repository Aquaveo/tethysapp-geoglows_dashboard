```python
import pandas as pd
import json
```

```python
data_folder = "workspaces/app_workspace/"
```

```python
country_file = open(data_folder + "ecuador.json")
country_data = json.load(country_file)
features = country_data["features"]
```

```python
precip_data = pd.read_csv(data_folder + "Ecuador_precip.csv")
precip_dict = dict()
for column in precip_data.columns[1:]:
    precip_dict[column] = precip_data[["month", column]].values.tolist()
```

```python
soil_data = pd.read_excel(data_folder + "Ecuador_soil.xlsx")
soil_data["month"] = soil_data["month"].astype("string")
soil_dict = dict()
for column in soil_data.columns[1:]:
    soil_dict[column] = soil_data[["month", column]].values.tolist()
```

```python
for feature in features:
   properties = feature["properties"]
   name = properties["ADM1_ES"]
   properties["precipitation"] = precip_dict[name]
   properties["soil moisture"] = soil_dict[name]
   print(properties)
```

```python
with open(f'{data_folder}/hydrosos_ecuador.json', 'w') as file:
    json.dump(country_data, file)
```
